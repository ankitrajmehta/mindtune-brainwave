# MindTune EEG Data Analysis

Analysis of EEG data collected using the MindTune collector UI.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
from datetime import datetime
import json

# Set style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('husl')
%matplotlib inline
```

## 1. Load All Sessions

```python
def load_all_sessions(sessions_dir='sessions'):
    """Load all EEG data from session folders (including nested)."""
    sessions_path = Path(sessions_dir)
    all_data = []
    
    # Find all eeg_rows.csv files recursively
    for eeg_file in sessions_path.rglob('eeg_rows.csv'):
        try:
            df = pd.read_csv(eeg_file)
            if len(df) > 1:  # Skip empty sessions
                df['session_folder'] = eeg_file.parent.name
                all_data.append(df)
        except Exception as e:
            print(f"Error loading {eeg_file}: {e}")
    
    combined = pd.concat(all_data, ignore_index=True)
    print(f"Loaded {len(all_data)} sessions with {len(combined)} total rows")
    return combined

df = load_all_sessions()
df.head()
```

```python
# Basic info
print("Columns:", df.columns.tolist())
print(f"\nShape: {df.shape}")
print(f"\nSessions: {df['session_id'].nunique()}")
df.describe()
```

## 2. Signal Quality Analysis

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

# Distribution of signal quality
axes[0].hist(df['signal_quality'], bins=50, edgecolor='black', alpha=0.7)
axes[0].set_xlabel('Signal Quality')
axes[0].set_ylabel('Frequency')
axes[0].set_title('Signal Quality Distribution\n(0=Good, 200=Not Worn)')
axes[0].axvline(x=0, color='green', linestyle='--', label='Good (0)')
axes[0].axvline(x=200, color='red', linestyle='--', label='Not worn (200)')
axes[0].legend()

# Signal quality by session
session_signal = df.groupby('session_id')['signal_quality'].mean().sort_values()
axes[1].barh(range(len(session_signal)), session_signal.values)
axes[1].set_xlabel('Mean Signal Quality')
axes[1].set_ylabel('Session')
axes[1].set_title('Mean Signal Quality by Session')

# Good signal percentage per session
good_signal_pct = df.groupby('session_id').apply(lambda x: (x['signal_quality'] == 0).mean() * 100)
axes[2].hist(good_signal_pct, bins=20, edgecolor='black', alpha=0.7, color='green')
axes[2].set_xlabel('% Good Signal (signal_quality=0)')
axes[2].set_ylabel('Number of Sessions')
axes[2].set_title('Good Signal Percentage Distribution')

plt.tight_layout()
plt.show()
```

## 3. Attention & Meditation Analysis

```python
# Filter for good signal data
df_good = df[df['signal_quality'] == 0].copy()
print(f"Good signal rows: {len(df_good)} ({len(df_good)/len(df)*100:.1f}%)")

fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Attention distribution
axes[0, 0].hist(df_good['attention'], bins=30, edgecolor='black', alpha=0.7, color='blue')
axes[0, 0].set_xlabel('Attention Level')
axes[0, 0].set_ylabel('Frequency')
axes[0, 0].set_title('Attention Distribution (0-100)')
axes[0, 0].axvline(df_good['attention'].mean(), color='red', linestyle='--', label=f'Mean: {df_good["attention"].mean():.1f}')
axes[0, 0].legend()

# Meditation distribution
axes[0, 1].hist(df_good['meditation'], bins=30, edgecolor='black', alpha=0.7, color='purple')
axes[0, 1].set_xlabel('Meditation Level')
axes[0, 1].set_ylabel('Frequency')
axes[0, 1].set_title('Meditation Distribution (0-100)')
axes[0, 1].axvline(df_good['meditation'].mean(), color='red', linestyle='--', label=f'Mean: {df_good["meditation"].mean():.1f}')
axes[0, 1].legend()

# Attention vs Meditation scatter
axes[1, 0].scatter(df_good['attention'], df_good['meditation'], alpha=0.3, s=10)
axes[1, 0].set_xlabel('Attention')
axes[1, 0].set_ylabel('Meditation')
axes[1, 0].set_title('Attention vs Meditation')
corr = df_good['attention'].corr(df_good['meditation'])
axes[1, 0].annotate(f'r = {corr:.3f}', xy=(0.05, 0.95), xycoords='axes fraction', fontsize=12)

# Attention & Meditation by session (boxplot)
session_means = df_good.groupby('session_id')[['attention', 'meditation']].mean()
session_means.plot(kind='box', ax=axes[1, 1])
axes[1, 1].set_ylabel('Level (0-100)')
axes[1, 1].set_title('Attention & Meditation Across Sessions')

plt.tight_layout()
plt.show()
```

## 4. Brainwave Band Analysis

```python
bands = ['delta', 'theta', 'low_alpha', 'high_alpha', 'low_beta', 'high_beta', 'low_gamma', 'mid_gamma']

fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.flatten()

for i, band in enumerate(bands):
    # Use log scale for better visualization
    data = np.log10(df_good[band].replace(0, np.nan).dropna())
    axes[i].hist(data, bins=30, edgecolor='black', alpha=0.7)
    axes[i].set_xlabel(f'log10({band})')
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'{band.replace("_", " ").title()}')

plt.suptitle('Brainwave Band Power Distributions (Log Scale)', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

```python
# Mean band power comparison
band_means = df_good[bands].mean()

fig, ax = plt.subplots(figsize=(10, 5))
colors = plt.cm.viridis(np.linspace(0, 1, len(bands)))
bars = ax.bar(range(len(bands)), np.log10(band_means), color=colors, edgecolor='black')
ax.set_xticks(range(len(bands)))
ax.set_xticklabels([b.replace('_', '\n') for b in bands], rotation=0)
ax.set_ylabel('Log10(Mean Power)')
ax.set_title('Mean Brainwave Band Power (Log Scale)')
plt.tight_layout()
plt.show()
```

## 5. Band Ratios (Useful Metrics)

```python
# Calculate useful band ratios
df_good['alpha'] = df_good['low_alpha'] + df_good['high_alpha']
df_good['beta'] = df_good['low_beta'] + df_good['high_beta']
df_good['gamma'] = df_good['low_gamma'] + df_good['mid_gamma']

# Common ratios used in neurofeedback
df_good['theta_beta_ratio'] = df_good['theta'] / (df_good['beta'] + 1)
df_good['alpha_theta_ratio'] = df_good['alpha'] / (df_good['theta'] + 1)
df_good['beta_alpha_ratio'] = df_good['beta'] / (df_good['alpha'] + 1)

ratios = ['theta_beta_ratio', 'alpha_theta_ratio', 'beta_alpha_ratio']

fig, axes = plt.subplots(1, 3, figsize=(15, 4))

for i, ratio in enumerate(ratios):
    data = df_good[ratio].clip(upper=df_good[ratio].quantile(0.99))  # Remove extreme outliers
    axes[i].hist(data, bins=50, edgecolor='black', alpha=0.7)
    axes[i].set_xlabel(ratio.replace('_', ' ').title())
    axes[i].set_ylabel('Frequency')
    axes[i].set_title(f'{ratio.replace("_", " ").title()} Distribution')

plt.tight_layout()
plt.show()
```

## 6. Event Impact Analysis

```python
events = ['ev_speaking', 'ev_question', 'ev_scolded', 'ev_praised', 'ev_qna', 'ev_tech_issue']

# Check event distribution
event_counts = df[events].sum()
print("Event occurrence counts:")
print(event_counts.sort_values(ascending=False))
```

```python
# Attention and Meditation during different events
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

metrics = ['attention', 'meditation']
for idx, metric in enumerate(metrics):
    event_means = []
    event_names = []
    
    for event in events:
        if df_good[event].sum() > 0:
            on_mean = df_good[df_good[event] == 1][metric].mean()
            off_mean = df_good[df_good[event] == 0][metric].mean()
            event_means.append([off_mean, on_mean])
            event_names.append(event.replace('ev_', ''))
    
    x = np.arange(len(event_names))
    width = 0.35
    
    axes[idx].bar(x - width/2, [m[0] for m in event_means], width, label='Event OFF', alpha=0.8)
    axes[idx].bar(x + width/2, [m[1] for m in event_means], width, label='Event ON', alpha=0.8)
    axes[idx].set_xticks(x)
    axes[idx].set_xticklabels(event_names, rotation=45, ha='right')
    axes[idx].set_ylabel(f'Mean {metric.title()}')
    axes[idx].set_title(f'{metric.title()} During Events')
    axes[idx].legend()

plt.tight_layout()
plt.show()
```

## 6b. Brainwave Changes During Events

```python
# Detailed brainwave analysis by event
bands = ['delta', 'theta', 'low_alpha', 'high_alpha', 'low_beta', 'high_beta', 'low_gamma', 'mid_gamma']

# Calculate mean band power during each event (ON vs OFF)
event_band_data = {}
for event in events:
    if df_good[event].sum() > 10:  # Need enough samples
        on_data = df_good[df_good[event] == 1][bands].mean()
        off_data = df_good[df_good[event] == 0][bands].mean()
        # Calculate percentage change
        pct_change = ((on_data - off_data) / off_data) * 100
        event_band_data[event.replace('ev_', '')] = pct_change

# Create DataFrame
event_band_df = pd.DataFrame(event_band_data).T
print("Brainwave Power % Change During Events (ON vs OFF):")
event_band_df.round(1)
```

```python
# Heatmap of brainwave changes during events
plt.figure(figsize=(12, 6))
sns.heatmap(event_band_df, annot=True, cmap='RdBu_r', center=0, fmt='.1f',
            cbar_kws={'label': '% Change from Baseline'})
plt.title('Brainwave Power Changes During Events\n(% change when event is ON vs OFF)')
plt.xlabel('Brainwave Band')
plt.ylabel('Event Type')
plt.tight_layout()
plt.show()
```

```python
# Grouped bar chart - absolute values by event
fig, axes = plt.subplots(2, 3, figsize=(16, 10))
axes = axes.flatten()

for idx, event in enumerate(events):
    if df_good[event].sum() > 10:
        on_means = np.log10(df_good[df_good[event] == 1][bands].mean() + 1)
        off_means = np.log10(df_good[df_good[event] == 0][bands].mean() + 1)
        
        x = np.arange(len(bands))
        width = 0.35
        
        axes[idx].bar(x - width/2, off_means, width, label='Event OFF', alpha=0.8, color='steelblue')
        axes[idx].bar(x + width/2, on_means, width, label='Event ON', alpha=0.8, color='coral')
        axes[idx].set_xticks(x)
        axes[idx].set_xticklabels([b.replace('_', '\n') for b in bands], rotation=0, fontsize=8)
        axes[idx].set_ylabel('Log10(Power)')
        axes[idx].set_title(f'{event.replace("ev_", "").title()}')
        axes[idx].legend(fontsize=8)

plt.suptitle('Brainwave Band Power by Event State', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

## 6c. Statistical Significance of Event Effects

```python
from scipy import stats

# T-test for each event and metric
metrics_to_test = ['attention', 'meditation'] + bands
significance_results = []

for event in events:
    if df_good[event].sum() > 10:
        for metric in metrics_to_test:
            on_values = df_good[df_good[event] == 1][metric]
            off_values = df_good[df_good[event] == 0][metric]
            
            if len(on_values) > 5 and len(off_values) > 5:
                t_stat, p_value = stats.ttest_ind(on_values, off_values)
                significance_results.append({
                    'Event': event.replace('ev_', ''),
                    'Metric': metric,
                    'Mean_OFF': off_values.mean(),
                    'Mean_ON': on_values.mean(),
                    'Diff': on_values.mean() - off_values.mean(),
                    't_stat': t_stat,
                    'p_value': p_value,
                    'Significant': 'Yes' if p_value < 0.05 else 'No'
                })

sig_df = pd.DataFrame(significance_results)
print("Significant differences (p < 0.05):")
sig_df[sig_df['Significant'] == 'Yes'][['Event', 'Metric', 'Diff', 'p_value']].round(4)
```

## 6d. Event Transition Analysis

```python
# Analyze how metrics change before/during/after events
def analyze_event_windows(data, event_col, metric_col, window_before=5, window_after=5):
    """Analyze metric changes around event transitions."""
    results = {'before': [], 'during': [], 'after': []}
    
    # Find event start indices
    event_diff = data[event_col].diff()
    starts = data[event_diff == 1].index.tolist()
    ends = data[event_diff == -1].index.tolist()
    
    for start_idx in starts:
        pos = data.index.get_loc(start_idx)
        if pos >= window_before:
            before_vals = data.iloc[pos-window_before:pos][metric_col].values
            results['before'].extend(before_vals)
    
    results['during'] = data[data[event_col] == 1][metric_col].values.tolist()
    
    for end_idx in ends:
        pos = data.index.get_loc(end_idx)
        if pos + window_after < len(data):
            after_vals = data.iloc[pos:pos+window_after][metric_col].values
            results['after'].extend(after_vals)
    
    return results

# Analyze for speaking event (most common)
if df_good['ev_speaking'].sum() > 50:
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    for idx, metric in enumerate(['attention', 'meditation']):
        windows = analyze_event_windows(df_good.reset_index(), 'ev_speaking', metric)
        
        positions = [1, 2, 3]
        data_to_plot = [windows['before'], windows['during'], windows['after']]
        
        bp = axes[idx].boxplot(data_to_plot, positions=positions, patch_artist=True)
        colors = ['lightblue', 'lightcoral', 'lightgreen']
        for patch, color in zip(bp['boxes'], colors):
            patch.set_facecolor(color)
        
        axes[idx].set_xticklabels(['Before', 'During', 'After'])
        axes[idx].set_ylabel(metric.title())
        axes[idx].set_title(f'{metric.title()} Around Speaking Events')
        
        # Add means
        means = [np.mean(d) for d in data_to_plot]
        axes[idx].scatter(positions, means, color='red', marker='D', s=50, zorder=5, label='Mean')
        axes[idx].legend()
    
    plt.suptitle('Metrics Before/During/After Speaking Events', fontsize=12)
    plt.tight_layout()
    plt.show()
```

## 6e. Band Ratios by Event

```python
# Calculate key ratios
df_good['alpha_total'] = df_good['low_alpha'] + df_good['high_alpha']
df_good['beta_total'] = df_good['low_beta'] + df_good['high_beta']
df_good['theta_beta'] = df_good['theta'] / (df_good['beta_total'] + 1)
df_good['alpha_beta'] = df_good['alpha_total'] / (df_good['beta_total'] + 1)
df_good['theta_alpha'] = df_good['theta'] / (df_good['alpha_total'] + 1)

ratios = ['theta_beta', 'alpha_beta', 'theta_alpha']
ratio_labels = ['Theta/Beta\n(Focus indicator)', 'Alpha/Beta\n(Relaxation)', 'Theta/Alpha\n(Drowsiness)']

# Compare ratios across events
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for idx, (ratio, label) in enumerate(zip(ratios, ratio_labels)):
    event_ratio_means = []
    event_labels = []
    
    for event in events:
        if df_good[event].sum() > 10:
            on_mean = df_good[df_good[event] == 1][ratio].clip(upper=df_good[ratio].quantile(0.95)).mean()
            off_mean = df_good[df_good[event] == 0][ratio].clip(upper=df_good[ratio].quantile(0.95)).mean()
            event_ratio_means.append([off_mean, on_mean])
            event_labels.append(event.replace('ev_', ''))
    
    x = np.arange(len(event_labels))
    width = 0.35
    
    axes[idx].bar(x - width/2, [m[0] for m in event_ratio_means], width, label='OFF', alpha=0.8)
    axes[idx].bar(x + width/2, [m[1] for m in event_ratio_means], width, label='ON', alpha=0.8)
    axes[idx].set_xticks(x)
    axes[idx].set_xticklabels(event_labels, rotation=45, ha='right')
    axes[idx].set_ylabel('Ratio Value')
    axes[idx].set_title(label)
    axes[idx].legend()

plt.suptitle('Brainwave Ratios During Events', fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```

## 6f. Summary Table: Event Effects

```python
# Create comprehensive summary table
summary_data = []

for event in events:
    if df_good[event].sum() > 10:
        on_data = df_good[df_good[event] == 1]
        off_data = df_good[df_good[event] == 0]
        
        row = {
            'Event': event.replace('ev_', '').title(),
            'N_samples': int(on_data.shape[0]),
            'Attention_ON': on_data['attention'].mean(),
            'Attention_OFF': off_data['attention'].mean(),
            'Attention_Change': on_data['attention'].mean() - off_data['attention'].mean(),
            'Meditation_ON': on_data['meditation'].mean(),
            'Meditation_OFF': off_data['meditation'].mean(),
            'Meditation_Change': on_data['meditation'].mean() - off_data['meditation'].mean(),
            'Theta_Beta_ON': on_data['theta_beta'].clip(upper=on_data['theta_beta'].quantile(0.95)).mean(),
            'Theta_Beta_OFF': off_data['theta_beta'].clip(upper=off_data['theta_beta'].quantile(0.95)).mean(),
        }
        summary_data.append(row)

summary_df = pd.DataFrame(summary_data)
print("Event Effects Summary:")
summary_df.round(2)
```

## 9. Time Series Analysis (Single Session)

```python
# Pick a session with good data
session_counts = df_good.groupby('session_id').size()
best_session = session_counts.idxmax()
print(f"Analyzing session: {best_session} ({session_counts[best_session]} rows)")

session_df = df_good[df_good['session_id'] == best_session].copy()
session_df['time_sec'] = (session_df['timestamp_ms'] - session_df['timestamp_ms'].min()) / 1000
```

```python
fig, axes = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

# Attention and Meditation over time
axes[0].plot(session_df['time_sec'], session_df['attention'], label='Attention', alpha=0.8)
axes[0].plot(session_df['time_sec'], session_df['meditation'], label='Meditation', alpha=0.8)
axes[0].set_ylabel('Level (0-100)')
axes[0].set_title(f'Session {best_session}: Attention & Meditation Over Time')
axes[0].legend()
axes[0].set_ylim(0, 100)

# Band powers over time (normalized)
for band in ['theta', 'alpha', 'beta']:
    if band == 'alpha':
        data = session_df['low_alpha'] + session_df['high_alpha']
    elif band == 'beta':
        data = session_df['low_beta'] + session_df['high_beta']
    else:
        data = session_df[band]
    axes[1].plot(session_df['time_sec'], np.log10(data + 1), label=band.title(), alpha=0.7)
axes[1].set_ylabel('Log10(Power)')
axes[1].set_title('Brainwave Bands Over Time')
axes[1].legend()

# Events over time
for i, event in enumerate(events):
    if session_df[event].sum() > 0:
        axes[2].fill_between(session_df['time_sec'], i, i + session_df[event] * 0.8, 
                            alpha=0.5, label=event.replace('ev_', ''))
axes[2].set_xlabel('Time (seconds)')
axes[2].set_ylabel('Events')
axes[2].set_title('Active Events Over Time')
axes[2].legend(loc='upper right')

plt.tight_layout()
plt.show()
```
